```python
import numpy as np
import matplotlib.pyplot as plt
pi = np.pi
```

```python

x = np.linspace(0,2*pi,100)

def negative(b):
    if b > 0:
        return -b
    else:
        return b
```

```python
b = 2
a = 5


y = a**(b*x)
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rc("font", size=18) 

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")

plt.savefig('CJ2P1',bbox_inches='tight',dpi=600)
```
